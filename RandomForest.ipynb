# Random Forest
Random Forest uses ***many trees***, and it makes a prediction by ***averaging*** the predictions of each component tree. It generally has much better than a single decision tree and it work well with default parameters.

It's in `sklearn.ensemble`

## Loading data

```python
import pandas as pd
melbourne_path = "./melb_data.csv"
mel_df = pd.read_csv(melbourne_path)
mel_df.dropna(axis=0)
```

```python
mel_df.columns
```

```python
y = mel_df['Price']

features = ['Rooms','Bathroom','Landsize','Lattitude','Longtitude']
X = mel_df[features]
```

## Train test split


```python
from sklearn.model_selection import train_test_split
train_X,val_X,train_y,val_y = train_test_split(X,y,random_state=0)
```

## Build model

```python
#Import 
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state=1)
model.fit(train_X,train_y)
```

```python
from sklearn.metrics import mean_absolute_error
pred_Y = model.predict(val_X)
mae = mean_absolute_error(pred_Y,val_y)
print(mae)
```

The optimized of decision tree has the MAE = 250000, but the Random Forest just 150000 -> much more better
